# Plotting Concrete Data

```python
from pandas import read_excel
```

```python
df = read_excel("./ENB2012_data.xlsx",sheet_name="Sheet1",header=0,nrows=769,dtype=float)
df
```

```python
df.loc[:, df.columns != 'Concrete compressive strength(MPa, megapascals) '].plot(
    kind="box",
    title="Box plots of each component and resulting concrete compressive strength",
    figsize=(20, 8),
)
```
